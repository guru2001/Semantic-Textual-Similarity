```python
snli = pd.read_csv('~/snli_1.0/snli_1.0_test.txt', sep = '\t')
```

```python
snli.drop(snli.columns.difference(['gold_label','sentence1', 'sentence2']), 1, inplace=True)
```

```python
snli.head()
```

```python
snli = snli[snli['gold_label'] != '-']
```

```python
for i, row in snli.iterrows():
    snli.at[i, 'input'] = " <cls> " + str(row[1]) + " <sep> " + str(row[2]) + " <cls> "

    if row[0] == 'neutral':
        snli.at[i, 'label'] = int(0)
    elif row[0] == 'entailment':
        snli.at[i, 'label'] = int(1)
    elif row[0] == 'contradiction':
        snli.at[i, 'label'] = int(2)
    else:
        print(row[0])
        break
```
